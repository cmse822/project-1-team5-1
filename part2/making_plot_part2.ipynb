```python
import subprocess
import matplotlib.pyplot as plt
from matplotlib.patches import PathPatch
from matplotlib.path import Path

def convert_ps_to_png(ps_file_path, png_file_path):
    # Use Ghostscript to convert PS to PNG
    command = [
        'gs',
        '-o', png_file_path,
        '-sDEVICE=pngalpha',
        '-r300',  # Set resolution (adjust as needed)
        ps_file_path
    ]
    subprocess.run(command)

def parse_png_and_plot(png_file_path):
    # Read the PNG file using Matplotlib
    img = plt.imread(png_file_path)

    # Create Matplotlib plot
    fig, ax = plt.subplots()
    ax.imshow(img)

    # Show the plot
    plt.show()

# Replace 'your_ps_file.ps' with the path to your PS file
ps_file_path = '/home/sonachitchyan/cs-roofline-toolkit/Empirical_Roofline_Tool-1.1.0/Results.RJ12/Run.001/roofline.ps'
png_file_path = 'output.pdf'

# Convert PS to PNG
convert_ps_to_png(ps_file_path, png_file_path)

# Parse PNG and create Matplotlib plot
parse_png_and_plot(png_file_path)
```
