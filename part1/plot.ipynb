```python
import matplotlib.pyplot as plt
```

Machine: Intel(R) Xeon(R) Gold 6148 CPU @ 2.40GHz (Intel18)   
Theoretical Peak = 96 Gflop/s    

```python
Ns = [1,10,30,80,160,280,420,610,830,1100,1420,1780,2200,3190]
Gflops = [float('inf'),0.155345,0.164125,0.16954,0.1681,0.167489,0.155973,0.144105,0.151699,0.147776,0.142116,0.13047,0.125803,0.101989]

plt.figure(figsize=(10, 6))
plt.plot(Ns, Gflops, label='Measured Gflop/s')
plt.plot(Ns, [96] * len(Ns), label='Theoretical Gflop/s')
plt.title('Intel18 Gflop/s vs N')
plt.xlabel('N')
plt.ylabel('Gflop/s')
plt.legend()
plt.show()
```
