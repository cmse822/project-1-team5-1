```python
import matplotlib.pyplot as plt
```

Machine: Intel(R) Xeon(R) Gold 6148 CPU @ 2.40GHz (Intel18)   
Theoretical Peak = 96 Gflop/s    

```python
Ns = [1,10,30,80,160,280,420,610,830,1100,1420,1780,2200,3190,4410,5110]
Gflops = [0.000932068,0.123362,0.155132,0.122928,0.124295,0.120966,0.110169,0.102688,0.106896,0.0994969,0.0882261,0.0728655,0.0727398,0.0703735,0.0701491,0.0724928]

plt.figure(figsize=(10, 6))
plt.plot(Ns, Gflops, label='Measured Gflop/s')
plt.plot(Ns, [96/40] * len(Ns), label='Theoretical Gflop/s')
plt.title('Intel18 Gflop/s vs N')
plt.xlabel('N')
plt.ylabel('Gflop/s')
plt.legend()
plt.show()
```
